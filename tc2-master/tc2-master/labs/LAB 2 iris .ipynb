```python
# Load the dataset
import pandas as pd

# Assuming the dataset is stored as iris.csv
df = pd.read_csv('iris.csv')

# Check the structure of the dataset
print(df.info())

# Summary statistics for numerical variables
print(df.describe())

# Unique values in categorical variable (Species)
print(df['Species'].unique())
```

```python
# Calculate the PMF for the Species variable
species_counts = df['Species'].value_counts(normalize=True)

# Visualize the PMF
import matplotlib.pyplot as plt

species_counts.plot(kind='bar')
plt.title('PMF of Species')
plt.xlabel('Species')
plt.ylabel('Probability')
plt.show()
```

```python
# Import seaborn for easy visualization
import seaborn as sns

# Visualize the PDF using seaborn's KDE plot
sns.kdeplot(df['SepalLengthCm'], shade=True)
plt.title('PDF of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np

# Sort the data for the CDF
sorted_sepal_length = np.sort(df['SepalLengthCm'])

# Compute the CDF
cdf = np.arange(1, len(sorted_sepal_length) + 1) / len(sorted_sepal_length)

# Plot the CDF
plt.plot(sorted_sepal_length, cdf)
plt.title('CDF of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Cumulative Probability')
plt.show()

```
