```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# 1. Uniform Distribution
# Problem Statement: Generate 1000 random numbers from a uniform distribution between 0 and 1.
# Create a histogram to visualize the distribution.

# Generate random numbers
uniform_data = np.random.uniform(0, 1, 1000)

# Create histogram
plt.hist(uniform_data, bins=30, color='blue', edgecolor='black')
plt.title("Uniform Distribution (0 to 1)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 2. Normal Distribution
# Problem Statement: Generate 1000 random numbers from a normal distribution with mean 0 and standard deviation 1.
# Create a histogram to visualize the distribution.

# Generate random numbers
normal_data = np.random.normal(0, 1, 1000)

# Create histogram
plt.hist(normal_data, bins=30, color='green', edgecolor='black')
plt.title("Normal Distribution (mean=0, std=1)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Exponential Distribution
# Problem Statement: Generate 1000 random numbers from an exponential distribution with a rate parameter of 0.5.
# Create a histogram to visualize the distribution.

# Generate random numbers
exponential_data = np.random.exponential(1 / 0.5, 1000)

# Create histogram
plt.hist(exponential_data, bins=30, color='red', edgecolor='black')
plt.title("Exponential Distribution (rate=0.5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Binomial Distribution
# Problem Statement: Simulate 1000 trials of a binomial distribution with 20 trials per experiment and a success probability of 0.3.
# Create a histogram to visualize the distribution.

# Generate random numbers
binomial_data = np.random.binomial(20, 0.3, 1000)

# Create histogram
plt.hist(binomial_data, bins=range(0, 21), color='purple', edgecolor='black', align='left')
plt.title("Binomial Distribution (n=20, p=0.3)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Poisson Distribution
# Problem Statement: Simulate 1000 events from a Poisson distribution with an average rate of 5 events per interval.
# Create a histogram to visualize the distribution.

# Generate random numbers
poisson_data = np.random.poisson(5, 1000)

# Create histogram
plt.hist(poisson_data, bins=range(0, max(poisson_data) + 1), color='orange', edgecolor='black', align='left')
plt.title("Poisson Distribution (lambda=5)")
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```
